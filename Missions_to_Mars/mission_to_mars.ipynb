## Scrape the Mars News Site and collect the latest News Title and Paragraph Text. Assign the text to variables that you can reference later. 

```python
# Dependencies
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pandas as pd

import time
```

```python
# setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless = False)
```

```python
# URL to be scraped
# calling on our browser and pass in the URL we want to scrape
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
time.sleep(1)
```

```python
html = browser.html 
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
soup.find('div', class_ = 'content_title')
```

```python
news_title = soup.find('div', class_ = 'content_title').get_text()
print(news_title)
```

```python
news_p = soup.find('div', class_ = 'article_teaser_body').get_text()
print(news_p)
```

## Mars Hemispheres

```python
url = 'https://marshemispheres.com/'

browser.visit(url)
```

```python
# create a list to hold images and titles
hemisphere_image_urls = []

# get the links of all the hemispheres
link = browser.find_by_css('a.product-item img')

# loop through the links, click the link, find th4e sample anchor, and return the image
for i in range(len(link)):
    
    # hemisphere info dictionary 
    hemisphereInfo = {}
    
    # find the elements on each loop to avoid a stale element exception
    browser.find_by_css('a.product-item img')[i].click()
    
    # find the sample image anchor tag and extract href
    sample = browser.links.find_by_text('Sample').first
    hemisphereInfo["img_url"] = sample['href']
    
    # get the hemisphere title
    hemisphereInfo['title'] = browser.find_by_css('h2.title').text
    
    # append hemisphere object to list
    hemisphere_image_urls.append(hemisphereInfo)
    
    # navigate backwards
    browser.back()
```
